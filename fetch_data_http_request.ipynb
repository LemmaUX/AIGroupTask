# Import Required Libraries
Import the necessary libraries, including requests and json.

```python
# Import the necessary libraries, including requests and json
import requests
import json
```

# Fetch Data from Public API
Write a Python script to perform an HTTP GET request to fetch JSON data from a public API.

```python
# Fetch Data from Public API

# Define the URL of the public API
url = "https://opentdb.com/api.php?amount=10"

# Perform an HTTP GET request to fetch the JSON data
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON data
    data = response.json()
    # Print the JSON data
    print(json.dumps(data, indent=4))
else:
    print(f"Failed to fetch data. HTTP Status code: {response.status_code}")
```

# Parse JSON Data
Parse the fetched JSON data into a Python dictionary.

```python
# Parse JSON Data

# Parse the fetched JSON data into a Python dictionary
parsed_data = json.loads(response.text)

# Print the parsed data to verify
print(parsed_data)
```

# Display Data
Display the parsed data in a readable format.

```python
# Display the fetched trivia questions
for question in parsed_data['results']:
    print(f"Category: {question['category']}")
    print(f"Type: {question['type']}")
    print(f"Difficulty: {question['difficulty']}")
    print(f"Question: {question['question']}")
    print(f"Correct Answer: {question['correct_answer']}")
    print(f"Incorrect Answers: {question['incorrect_answers']}")
    print("\n")
```
