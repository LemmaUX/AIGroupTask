### Format and print the filtered titles using f-strings.

```python
# Import Required Libraries
import requests
import json

# Fetch Data from Public API
url = "https://opentdb.com/api.php?amount=10"
response = requests.get(url)

# Parse JSON Data
parsed_data = response.json()

# Check if 'results' key exists in the parsed data
if 'results' in parsed_data:
    # Display the fetched trivia questions
    for question in parsed_data['results']:
        print(f"Category: {question['category']}")
        print(f"Type: {question['type']}")
        print(f"Difficulty: {question['difficulty']}")
        print(f"Question: {question['question']}")
        print(f"Correct Answer: {question['correct_answer']}")
        print(f"Incorrect Answers: {question['incorrect_answers']}")
        print("\n")

    # Process Data with Conditional List Comprehension
    # Extract questions that contain the string 'qui'
    filtered_questions = [question['question'] for question in parsed_data['results'] if 'qui' in question['question']]

    # Display Filtered Questions with f-Strings
    for title in filtered_questions:
        print(f"Filtered Question: {title}")

    # Type Checking using List Comprehension
    # Check if any of the extracted titles is of type int
    type_checks = [f"The title '{title}' is of type int." if isinstance(title, int) else f"The title '{title}' is of type {type(title).__name__}." for title in filtered_questions]

    # Display Type Checks
    for check in type_checks:
        print(check)
else:
    print("The key 'results' was not found in the JSON response.")
```
